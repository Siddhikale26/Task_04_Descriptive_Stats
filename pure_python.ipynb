```python
import csv
import math
from collections import defaultdict, Counter
```

```python
filename = "/Users/siddhikale/Downloads/period_03/2024_fb_ads_president_scored_anon.csv"  # Replace with actual file path

with open(filename, newline='', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    rows = list(reader)
    columns = reader.fieldnames

print(f"{len(rows)} rows with {len(columns)} columns.")

```

```python
def is_float(val):
    try:
        float(val)
        return True
    except:
        return False

numeric_cols = []
categorical_cols = []
for col in columns:
    sample = [row[col] for row in rows[:100] if row[col] != '']
    if all(is_float(v) for v in sample):
        numeric_cols.append(col)
    else:
        categorical_cols.append(col)

print("Numeric Columns:")
for col in numeric_cols:
    print(f"  - {col}")

print("\nCategorical Columns:")
for col in categorical_cols:
    print(f"  - {col}")
```

```python
import math
from collections import Counter

def is_float(val):
    try:
        float(val)
        return True
    except:
        return False

def compute_numeric_stats(values):
    values = [float(v) for v in values if is_float(v)]
    n = len(values)
    if n == 0:
        return {}
    mean = sum(values) / n
    min_val = min(values)
    max_val = max(values)
    std_dev = math.sqrt(sum((x - mean) ** 2 for x in values) / n)
    return {
        'Count': n,
        'Mean': round(mean, 4),
        'Min': round(min_val, 4),
        'Max': round(max_val, 4),
        'Std Dev': round(std_dev, 4)
    }

def compute_categorical_stats(values):
    freq = Counter(values)
    most_common = freq.most_common(1)[0] if freq else (None, 0)
    return {
        'Unique Values': len(freq),
        'Most Common': most_common[0],
        'Frequency': most_common[1]
    }

print("\n Numeric Columns Summary:\n")
for col in numeric_cols:
    vals = [row[col] for row in rows if row[col] != '']
    stats = compute_numeric_stats(vals)
    print(f"{col}:")
    for k, v in stats.items():
        print(f"    {k}: {v}")
    print()

print("\n Categorical Columns Summary:\n")
for col in categorical_cols:
    vals = [row[col] for row in rows if row[col] != '']
    stats = compute_categorical_stats(vals)
    print(f"{col}:")
    for k, v in stats.items():
        print(f"    {k}: {v}")
    print()
```

```python
def group_by(data, keys):
    grouped = defaultdict(list)
    for row in data:
        key = tuple(row[k] for k in keys)
        grouped[key].append(row)
    return grouped

# Group by page_id
page_groups = group_by(rows, ['page_id'])
for pid, group_rows in list(page_groups.items())[:2]:
    print(f"\nGroup: page_id = {pid}")
    for col in numeric_cols:
        vals = [row[col] for row in group_rows if row[col] != '']
        print(f"  {col}: {compute_numeric_stats(vals)}")

# Group by page_id + ad_id
page_ad_groups = group_by(rows, ['page_id', 'ad_id'])
for key, group_rows in list(page_ad_groups.items())[:2]:
    print(f"\nGroup: page_id = {key[0]}, ad_id = {key[1]}")
    for col in numeric_cols:
        vals = [row[col] for row in group_rows if row[col] != '']
        print(f"  {col}: {compute_numeric_stats(vals)}")
```
