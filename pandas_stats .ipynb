# Script 2 — Descriptive Stats Using Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/siddhikale/Downloads/period_03/2024_fb_ads_president_scored_anon.csv')  # Update path if needed
df.head()
```

```python
df.describe(include='all').T
```

```python
# Show unique count and top values for object columns
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print("Unique Values:", df[col].nunique())
    print("Top Values:\n", df[col].value_counts().head(3))
```

```python
# Group by page_id
grouped_page = df.groupby('page_id').describe()
grouped_page.head()
```

```python
# Group by page_id and ad_id
grouped_page_ad = df.groupby(['page_id', 'ad_id']).describe()
grouped_page_ad.head()
```
