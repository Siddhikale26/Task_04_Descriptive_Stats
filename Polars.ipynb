```python
import polars as pl
```

```python
df = pl.read_csv('/Users/siddhikale/Downloads/period_03/2024_fb_ads_president_scored_anon.csv')
df.head()
```

```python
df.describe()
```

```python

categorical_cols = [col for col in df.columns if df[col].dtype == pl.Utf8]

for col in categorical_cols:
    print(f"\n📌 Column: {col}")
    
    # Get number of unique values
    unique = df.select(pl.col(col).n_unique()).to_series()[0]
    print("Unique values:", unique)
    
    # Get top 3 most frequent values
    top_3 = (
        df.group_by(col)
          .agg(pl.count().alias("count"))
          .sort("count", descending=True)
          .limit(3)
    )
    print("Top 3:\n", top_3)
```

```python
# Group by page_id
grouped_page = df.group_by("page_id").agg([
    pl.col(col).mean().alias(f"{col}_mean") 
    for col in df.columns 
    if df[col].dtype in [pl.Int64, pl.Float64]
])
grouped_page.head()
```

```python
# Group by page_id and ad_id
grouped_page_ad = df.group_by(["page_id", "ad_id"]).agg([
    pl.col(col).mean().alias(f"{col}_mean") 
    for col in df.columns 
    if df[col].dtype in [pl.Int64, pl.Float64]
])
grouped_page_ad.head()
```
